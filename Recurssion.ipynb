```python
# Print 1-n from using the recurssion

def fun1(n):
    if n<=0:
        return 
    fun1(n-1)
    print(n)
```

```python
fun1(5)
```

```python
# print n-1 from using the recurssion

def fun2(n):
    if n<=0:
        return
    print(n)
    fun2(n-1)
```

```python
fun2(5)
```

```python
# print the value from interger to binary

def fun3(n):
    if n==0:
        return
    fun3(n/2)
    print(n%2)
```

```python
fun3(6)
```

```python
# print the output of the given number

def fun4(n):
    if n==0:
        return
    else:
        return 1 + fun4(n/2)
```

```python
print(fun4(5))
```

```python
# print the sum of digit

def fun5(n):
    
    if n<=10:
        return
    return fun5(n//10)+n%10
    
```

```python
fun5(123)
```
